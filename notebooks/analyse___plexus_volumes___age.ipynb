```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats import f_oneway, kruskal, levene, ttest_ind, pearsonr, spearmanr
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
# Load the data
patho_data = pd.read_csv("../Dataset002_DFGFinetuned_2d_full_pathologic___volumes.csv")
patho_metadata = pd.read_excel("../MSData_noNMO.xlsx")
hc_data = pd.read_csv("../Dataset002_DFGFinetuned_2d_full_hc___volumes.csv")
hc_metadata = pd.read_excel("../HCData.xlsx")

# Adjust the column name for the patient ID (replace 'actual_patient_id_column' with the correct name)
patho_metadata_id_column = "patID"
hc_metadata_id_column = "subID"
data_id_column = "subject_id"
```

```python
# Merge the volume data with metadata
patho_merged_data = patho_metadata.merge(patho_data[['subject_id', 'volume']], left_on=patho_metadata_id_column, right_on=data_id_column)
hc_merged_data = hc_metadata.merge(hc_data[['subject_id', 'volume']], left_on=hc_metadata_id_column, right_on=data_id_column)
merged_data = pd.concat([patho_merged_data, hc_merged_data], axis=0)
```

# Pathologic data (MS)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=patho_merged_data, x='Alter', y='volume', alpha=0.7)
sns.regplot(data=patho_merged_data, x='Alter', y='volume', scatter=False, color='red')
plt.title("Plexus Volume vs. Age (MS)")
plt.xlabel("Age (Years)")
plt.ylabel("Choroid Plexus Volume")
plt.savefig("volume_by_age_ms.png")
plt.show()
```

```python
# Fit regression model
model = smf.ols(formula="volume ~ Alter", data=patho_merged_data).fit()

# Print the summary
print(model.summary())
```

```python
# Pearson correlation (linear relationship)
pearson_corr, pearson_p = pearsonr(patho_merged_data['Alter'], patho_merged_data['volume'])
print(f"Pearson Correlation: {pearson_corr:.3f}, p-value: {pearson_p:.3f}")

# Spearman correlation (non-parametric)
spearman_corr, spearman_p = spearmanr(patho_merged_data['Alter'], patho_merged_data['volume'])
print(f"Spearman Correlation: {spearman_corr:.3f}, p-value: {spearman_p:.3f}")
```
