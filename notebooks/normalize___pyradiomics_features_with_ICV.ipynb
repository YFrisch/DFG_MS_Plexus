```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# df_rad_path = Path('/media/yannik/Intenso/DATA/dfg_plexus/radiomics_features___combined_SA.csv')
df_rad_path = Path('/mnt/Intenso/DATA/dfg_plexus/radiomics_features___harmonized___combined___SA.csv')
assert(df_rad_path.exists())
df_rad = pd.read_csv(df_rad_path, delimiter=';')
df_rad_columns = df_rad.columns.tolist()

df_icv_path = Path('/mnt/Intenso/DATA/dfg_plexus/volumes_normalized.xlsx')
assert(df_icv_path.exists())
df_icv = pd.read_excel(df_icv_path)
df_icv_subset = df_icv[['Volumes', 'Total-ICV']]
```

```python
df = pd.merge(df_rad, df_icv_subset, left_on='patID', right_on='Volumes', how='inner')
#print(df.head())
print(df_rad.shape)
print(df.shape)
```

```python
volumetric_features = [
    'original_shape_VoxelVolume',
    'original_shape_MeshVolume'
]
area_features = [
    'original_shape_SurfaceArea'
]
linear_features = [
    'original_shape_Maximum2DDiameterColumn',
    'original_shape_Maximum2DDiameterRow',
    'original_shape_Maximum2DDiameterSlice',
    'original_shape_Maximum3DDiameter',
    'original_shape_MajorAxisLength',
    'original_shape_MinorAxisLength',
    'original_shape_LeastAxisLength'
]
```

```python
# Volume (ICV^1)
for feat in volumetric_features:
    if feat in df.columns:
        df[feat] = df[feat] / df['Total-ICV']

# Area (ICV^(2/3))
for feat in area_features:
    if feat in df.columns:
        df[feat] = df[feat] / (df['Total-ICV'] ** (2/3))

# Length (ICV^(1/3))
for feat in linear_features:
    if feat in df.columns:
        df[feat] = df[feat] / (df['Total-ICV'] ** (1/3))
```

```python
df_final = df[df_rad_columns]
# df_final.to_csv(df_rad_path.parents[0] / 'radiomics_features___combined_SA___normalized.csv', index=False, sep=';')
df_final.to_csv(df_rad_path.parents[0] / 'radiomics_features___harmonized___combined___SA___normalized_ICV.csv', index=False, sep=';')
```
